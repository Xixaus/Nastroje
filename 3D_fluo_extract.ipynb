```python
#@title Vstupn√≠ hodnoty { run: "auto", vertical-output: true }
file_name = "3D.XLS" #@param {type:"string"}
name = "vit B2" #@param {type:"string"}

treshold =  0 #@param {type:"number"}
sqrt = None #@param {type:"number"}

display_dataframes = True #@param {type:"boolean"}
reduction = False #@param {type:"boolean"}
histogram = True #@param {type:"boolean"}
plot_2D = True #@param {type:"boolean"}
plot_3D = True #@param {type:"boolean"}
```

```python
#@title Naƒçten√≠ a zpracov√°n√≠ dat

import pandas as pd
import numpy as np
import plotly.graph_objects as go

def data_extract(df, file_name, reduction = False):
    df1 = pd.DataFrame()

    if df.iloc[-1].isna().all():
        df = df.iloc[0:-1]

    NaN_rows = df[df.isnull().all(axis=1)].index.tolist() + [df.index.max()]

    x = 0

    for i in NaN_rows:
        dff = df.loc[x:i]
        dff = dff.dropna(axis=0, how="all").dropna(axis=1, how="all")

        cols_name = dff[1]
        rows = dff.iloc[0, list(range(0, len(dff.columns), 3))].tolist()
        cols = list(range(2, len(dff.columns), 3))

        dff = dff.iloc[:, cols]

        dff.set_index(cols_name, inplace=True)
        dff.columns = rows

        df1 = pd.concat([df1, dff], axis=1)
        x = i+1


    df_red = df1.iloc[::10]

    df1.to_csv(f"/content/{file_name[0:-4]}.csv")
    df_red.to_csv(f"/content/{file_name[0:-4]}_reduction.csv")

    if reduction == True:
        return(df_red)
    else:
        return(df1)

df = pd.read_excel(f"/content/{file_name}", header=None)
data = data_extract(df, file_name, reduction)

if display_dataframes == True:
    display(df)
    display(data)
```

```python
#@title Vykreslen√≠ graf≈Ø

x = data.columns.astype(float)
y = data.index
X, Y = np.meshgrid(x, y)

Z = data.values
Z = np.where(Z < treshold , 0, Z)

if sqrt != None:
    Z = Z ** (1/sqrt)

if histogram == True:
    s = data.melt()
    fig1 = go.Figure(data=[go.Histogram(x=s['value'], nbinsx=10000)])
    fig1.show()

if plot_2D == True:
    fig2 = go.Figure(data=go.Contour(z=Z,
                                     x=x, y=y
                                     ))

    fig2.update_layout(
        title=name,
        xaxis_title='EX [nm]',
        yaxis_title='EM [nm]',
        autosize=False,
        width=1000, height=1000
    )

    fig2.show()

if plot_3D == True:
    fig3 = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

    fig3.update_layout(title=name, autosize=True,
                    scene=dict(
                        xaxis_title='EX [nm]',
                        yaxis_title='EM [nm]',
                        zaxis_title='Intezita fluorescence',
                        aspectmode='cube'
                        ),
                    width=1000, height=1000,
                    )

    fig3.show()
```
